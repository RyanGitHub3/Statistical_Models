Below I used a 6 fold cross validation method in order to understand the balance that exists between Bias and Variance in my model, and plotted it below. I used an average of each result for each of the five sizes of training data, and compiled that into a table for both training and test data. 

```python
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression

train_sizes = [1,20,80,150,277]
features = ['P/E ratio','P/B ratio','Debt/Equity','EPS','Revenue Growth','EV/Sales','EV/FCF', 'Current Ratio','Free Cash Flow per Share','Shareholders Equity per Share']
target = 'Price'
train_sizes, train_scores, test_scores = learning_curve(estimator = LinearRegression(),X = Stock_Data[features],y = Stock_Data[target], 
                                                              train_sizes = train_sizes, cv = 6,scoring = 'neg_mean_squared_error')

train_scores_mean = -train_scores.mean(axis = 1)
test_scores_mean = -test_scores.mean(axis = 1)
print('Mean training scores\n\n', pd.Series(train_scores_mean, index = train_sizes))
print('\nMean testing scores\n\n',pd.Series(test_scores_mean, index = train_sizes))

plt.style.use('default')
plt.plot(train_sizes, train_scores_mean, label = 'Training error')
plt.plot(train_sizes, test_scores_mean, label = 'Test error')
plt.ylabel('MSE', fontsize = 12)
plt.xlabel('Training set size', fontsize = 12)
plt.title('Learning curves for Stock Price Valuation Model', fontsize = 13)
plt.legend()
plt.ylim(0,15000)
plt.xlim(50,300)
```

Above there is a high degree of variance from the training error and the validation error, roughly 3,000 units of error. Since I have a high variance and relatively low bias scenario, it is likely that the model is overfitting, and is in need of some type of feature reduction. A small degree of increased bias as a tradeoff to combat the variance in this case would be a satisfactory trade.

In order to do feature selection, I will implement Lasso Regression. To reduce the complexity of the model, I will assign my own alpha level of 50, in order to induce more sparsity than would typically be seen with a lower alpha. 

```python
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV, Ridge, Lasso, ElasticNet
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

X = Stock_Data[['P/E ratio','P/B ratio','Debt/Equity','EPS','Revenue Growth','EV/Sales','EV/FCF', 'Current Ratio','Free Cash Flow per Share','Shareholders Equity per Share']]
y = Stock_Data[['Price']]
y = y.values.ravel()


reg = Lasso(alpha = 50)
reg.fit(X, y)
coef = pd.Series(reg.coef_, index = X.columns)
ordered_coef = coef.sort_values()
print(ordered_coef[::-1])


ordered_coef.plot(kind = "barh")
plt.title("Bar Graph of Selected Variables")
plt.ylabel("Variable List")
```

How this Model can be used:
    
The main appeal of the model that I have created, is that with a little modification, it can be implemented as a major time saver for every investor. One of the most time consuming and difficult things in stock valuation, is reading all of the 10K and 10Q forms that are associated with each and every company, in order to figure out which ones are overvalued and which ones are undervalued. The problem with this, is that it may take many hours to determine if that company is worth digging into deeper and learning more about. 

With some small modifications, this model can be fed enormous quantities of data from thousands of different companies, and it can reasonably predict what the share price of that company should be versus what the share price of that company actually is. If the difference between the two prices is noteworthy, maybe something like a 35% difference between share price and predicted price, the model can save the name of that stock in a file for the investor to look at later on. In a matter of just a few days, the model would be able to sort through the amount of stocks that would take the typical investor years to do by hand, and it can compile a list of undervalued stocks that it believes an investor should look into further.

Drawbacks of this model:
    
The biggest drawback of this model, is that it does not tell the full story of a stock price. The model only uses 4 variables to capture the price. This is an unrealistic assumption, and in in reality there are often many things that go into what the price of a stock will be. Due to this, it is important that the investor does not just take the result of the model as a difinitive truth, but rather they use the model as a tool in order to find undervalued investments, and then research them further to make sure the balance sheet is in order, and either validate or disprove what the model predicted. Things like debt, lawsuits, and high growth expectations can all dramatically impact the price of a stock in ways that this model will not be able to capture.
